```python
#Create list of word and freq respectively
word = []
freq = []
n = []
n_freq = []
i = 1

#take values and key from dict and sort.
for k, v in sorted(dict.items(), key=lambda x: x[1], reverse=True):
    word.append(k)
    freq.append(v)
    n.append(i)
    n_freq.append(i*v)
    
    i += 1
    
```

```python
df = pd.DataFrame({'word' : word,
                                    'freq' : freq,
                                    'n': n,
                                    'n*freq': n_freq})
```
