```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Graph = nx.Graph()
```

```python
#Load file
f = open('alchohol.txt', 'r')
data = f.read()
f.close()

lines = data.split(' ') 
```

```python
lines
```

```python
c = [(lines[i],lines[i+1]) for i in range(0,len(lines)-1,1)]
print (c)
```

```python
Graph.add_edges_from(c)
```

```python
print (nx.info(Graph))
```

```python
nx.draw_networkx (Graph)
plt.savefig("pre_literature.png")
plt.show()
```

```python
#computing the mesures of network
print('the clustering coefficient for nodes = ',nx.clustering(Graph))
print ('\n')
print('degeree sequence = ',Graph.degree().values())
print ('\n')
print('degree histogram = ', nx.degree_histogram(Graph)) #Return a list of the frequency of each degree value.
print ('\n')
print('degree centrality = ',nx.degree_centrality(Graph)) 
print ('\n')
print('betweenness centrality = ', nx.betweenness_centrality(Graph)) 
```

```python
d = []
for i in range(0,len(lines)-1,1):
    for j in range(i+1,len(lines),1):
        d.append((lines[i], lines[j]))
```

```python
d
```

```python
Graph2 = nx.Graph()
Graph2.add_edges_from(d)
```
