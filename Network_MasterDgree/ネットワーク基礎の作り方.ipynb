```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Graph = nx.Graph()
```

```python
# https://networkx.github.io/documentation/networkx-1.10/reference/generated/networkx.generators.random_graphs.gnp_random_graph.html
H = nx.gnp_random_graph(7,0.5)

```

```python
Graph.add_edges_from(H.edges())
```

```python
nx.draw_networkx(H)
plt.show()
```

```python
G=nx.Graph()
G.add_edges_from([(1,2),(1,3)])
```

```python
nx.draw_networkx(G)
plt.show()
```
